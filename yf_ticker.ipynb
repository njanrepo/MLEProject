```python
def get_sp500_list():
  payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
  first_table = payload[0]
  sp500df = first_table
  sp500_list = sp500df.Symbol.values.tolist()
  return sp500_list
```

```python
sp_list = get_sp500_list()
s_info = []
i =0
for t in sp_list:
  print(t)
  i +=1
  print(i)
  inn = yf.Ticker(t).info
  s_info.append(inn)
```
